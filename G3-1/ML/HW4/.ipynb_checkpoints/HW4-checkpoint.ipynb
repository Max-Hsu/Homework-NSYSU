```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as skl
import sklearn.metrics as skm
import sklearn.preprocessing as skp
import seaborn as sns
csv = pd.read_csv('HW4_house-prices.csv') #requirement 1 reading the csv data

#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)
#pd.set_option('display.width', None)
#pd.set_option('display.max_colwidth', -1)

csv
```

```python
csv_dummies = pd.get_dummies(csv)　# requirement2 : one hot coding for non number type
```

```python
nul_dataframe = pd.isna(csv_dummies) # requirement3 : filling Nan with column average
for series_null in nul_dataframe:
    ser = nul_dataframe[series_null]
    if(len(np.where(ser)[0])):       # find Nan
        csv_dummies[str(series_null)].fillna(value=csv_dummies[str(series_null)].mean(),inplace=True) # fillna with average
```

```python
sale_price = csv_dummies['SalePrice'] # extract SalePrice
```

```python
csv_dummies = csv_dummies.drop(columns=['Id']) # remove Id from dataset , because it will affect the result I think
```

```python
y_train , y_test = sale_price[:round(len(sale_price)*0.7)] , sale_price[round(len(sale_price)*0.7):]
x_train , x_test = csv_dummies[:round(len(sale_price)*0.7)] , csv_dummies[round(len(sale_price)*0.7):]
y_test = np.array(y_test) #splitting data set with ratio of 7:3 , for the first part of requirement4
```

```python
linear_learn = skl.LinearRegression().fit(x_train,y_train)
ridge_learn = skl.Ridge().fit(x_train,y_train) 
lasso_learn = skl.Lasso().fit(x_train,y_train) #create three model for training
```

```python
linear_y_predict = linear_learn.predict(x_test)#predict the trained model with test data
ridge_y_predict = ridge_learn.predict(x_test)
lasso_y_predict = lasso_learn.predict(x_test)

MSE_list = []

linear_RSE = skm.mean_squared_error(y_test,linear_y_predict) # acquire the MSE
ridge_RSE = skm.mean_squared_error(y_test,ridge_y_predict)
lasso_RSE = skm.mean_squared_error(y_test,lasso_y_predict)

MSE_list.append((linear_RSE,'Linear Regression')) # append MSE to list in order to get min MSE
MSE_list.append((ridge_RSE,'Ridge Regression'))
MSE_list.append((lasso_RSE,'Lasso Regression'))

MSE_list.sort() # sort the list

print("Linear Regression's MSE: ",linear_RSE,"\n"\
      "Ridge Regression's MSE:  ",ridge_RSE,"\n"\
      "Lasso Regression's MSE:  ",lasso_RSE)

print("\n\n")
print(MSE_list[0][1]," is better in MSE.")
print("\n\n")
plt.figure()
_,plot = plt.subplots(1,3,figsize=(20,10))
plot[0].plot(y_test,linear_y_predict,'p')
plot[0].title.set_text('Linear Regression')
plot[0].set_ylabel('Predict House Price')
plot[0].set_xlabel('True House Price')
plot[1].plot(y_test,ridge_y_predict,'p')
plot[1].title.set_text('Ridge Regression')
plot[1].set_ylabel('Predict House Price')
plot[1].set_xlabel('True House Price')
plot[2].plot(y_test,lasso_y_predict,'p')
plot[2].title.set_text('Lasso Regression')
plot[2].set_ylabel('Predict House Price')
plot[2].set_xlabel('True House Price')
```
